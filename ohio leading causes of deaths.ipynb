```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cause='Chronic Diseases','Unintentional Injuries','Alzhemers','Influenza and Pneumonia','Sepsis','Others'
percentile=[62,5,4,2,1,26]
```

```python
plt.figure(figsize=(10,10))
explode=(0.05,0,0,0,0,0)
plt.pie(percentile,labels=cause,explode=explode,autopct='%1.1f%%',startangle=70)
plt.axis('equal')
plt.title('Ohio State:Leading causes of death')
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.bar(cause,percentile,width=0.4,color="green")
plt.title("Ohio leading causes of death")
plt.xlabel("Cause")
plt.ylabel("Percentage")
```

```python
plt.figure(figsize=(20,5))
plt.scatter(cause,percentile,marker="s",edgecolor="red")
plt.title("Ohio leading causes of death")
plt.xlabel("Cause")
plt.ylabel("Percentage")
```

```python
plt.figure(figsize=(20,5))
plt.plot(cause,percentile,color="red")
plt.title("Ohio leading causes of death")
plt.xlabel("Cause")
plt.ylabel("Percentage")
```

```python
cause='CD','UI','A','I P','S','O'
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
ax1.scatter(cause,percentile,marker="s",edgecolor="red")
ax2.plot(cause,percentile,color="red")
ax3.bar(cause,percentile,width=0.4,color="green")
ax4.pie(percentile,labels=cause,explode=explode,autopct='%1.1f%%',startangle=70)
ax1.title.set_text("Linear")
ax2.title.set_text("Double")
ax3.title.set_text("Square")
ax4.title.set_text("Cube")
fig.tight_layout()
```
